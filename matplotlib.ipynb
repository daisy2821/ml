```
hd=pd.read_csv("heartdisease.csv")
hd.head()
```

```
hd["age"].plot.hist(bins=10);
```

```
hd.plot.hist(figsize=(10,30),subplots=True);
```

```
over50=hd[hd["age"]>50]
over50.head()
```

```
over50.plot(kind="scatter",x="age",y="chol",c="target",figsize=(10,5));
```

```
fig,ax=plt.subplots(figsize=(10,6))
over50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax);
```

```
##oo from scratch
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over50["age"],y=over50["chol"],c=over50["target"]);
ax.set(title="herat disease and cholestrol level",xlabel="age",ylabel="cholestrol");
ax.legend(*scatter.legend_elements(),title="Target");
ax.axhline(over50["chol"].mean(),linestyle="--");
```

```
over50.head()
```

```
#subplot of chol ,age,thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True);
#add data
scatter=ax0.scatter(x=over50["age"],y=over50["chol"],c=over50["target"],cmap="winter")
#customize
ax0.set(title="heart disease nd cholestrol",ylabel="cholestrol");
#change x axis lim
ax0.set_xlim([50,80])
ax0.set_ylim([60,200])
#add legend
ax0.legend(*scatter.legend_elements(),title="Target");
#add mean line
ax0.axhline(over50["chol"].mean(),linestyle="--");
#aadd data to ax1
scatter=ax1.scatter(x=over50["age"],y=over50["thalach"],c=over50["target"],cmap="winter")
#customize
ax1.set(title="heart disease nd heart rate",xlabel="age",ylabel="Thalach");
#change x axis lim
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
#add legend
ax1.legend(*scatter.legend_elements(),title="Target");
#add mean line
ax1.axhline(over50["thalach"].mean(),linestyle="--");
#add title the figure
fig.suptitle("HEART DISEASE ANALYSIS",fontsize=16,fontweight="bold",fontstyle="oblique");
```
